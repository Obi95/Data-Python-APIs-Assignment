```python
# We only want the Driver Standings
for x in range(5):
    data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'][x]["Driver"]["familyName"]
    print(data)
my_racer_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
my_racer_data
```

```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"
        
        racer_dict[racer_name]={
            'first_name':racer['Driver']['givenName'],
            'last_name':racer['Driver']['familyName'],
            'position':racer['position'],
            'wins':racer['wins'],
            'DOB':racer['Driver']['dateOfBirth'],
            'nationality':racer['Driver']['nationality'],
            'constructor':racer['Constructors'][0]['name']
        }
        new_data.append(racer_dict)
    return new_data

get_racer_info(my_racer_data)
```

```python
import requests

def get_driver_info_by_year_and_round(year, r):
    url = f"http://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    response = requests.get(url)
    if not response.json()["MRData"]["StandingsTable"]["StandingsLists"]:
        return "We had an error loading your data it is likely the list or the round is not in the database"
    data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
    return get_racer_info(data)

get_driver_info_by_year_and_round(2019, 4)
```

```python
import requests

def get_pokemon(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        abilities = [ability["ability"]["name"] for ability in data["abilities"]]
        weight = data["weight"]
        return abilities, weight
    else:
        print(f"Error fetching data for {pokemon_name}: {response.status_code}")
        return None, None

def main():
    pokemon_names = [
        "charizard", "gengar", "onix", "pikachu", "bulbasaur", "squirtle", "jigglypuff", "meowth", 
        "psyduck", "machop", "magnemite", "geodude", "gastly", "haunter", "hypno", "cubone", 
        "eevee", "snorlax", "dragonite", "mewtwo"
    ]

    pokemon_data_by_type = {}

    for pokemon_name in pokemon_names:
        abilities, weight = get_pokemon(pokemon_name)
        if abilities and weight:
            if pokemon_name == "charizard":
                pokemon_type = "Fire"
            elif pokemon_name == "gengar":
                pokemon_type = "Ghost / Poison"
            elif pokemon_name == "onix":
                pokemon_type = "Rock / Ground"
            elif pokemon_name == "pikachu":
                pokemon_type = "Electric"
            elif pokemon_name == "bulbasaur":
                pokemon_type = "Grass / Poison"
            elif pokemon_name == "squirtle":
                pokemon_type = "Water"
            elif pokemon_name == "jigglypuff":
                pokemon_type = "Normal / Fairy"
            elif pokemon_name == "meowth":
                pokemon_type = "Normal"
            elif pokemon_name == "psyduck":
                pokemon_type = "Water"
            elif pokemon_name == "machop":
                pokemon_type = "Fighting"
            elif pokemon_name == "magnemite":
                pokemon_type = "Electric / Steel"
            elif pokemon_name == "geodude":
                pokemon_type = "Rock / Ground"
            elif pokemon_name == "gastly":
                pokemon_type = "Ghost / Poison"
            elif pokemon_name == "haunter":
                pokemon_type = "Ghost / Poison"
            elif pokemon_name == "hypno":
                pokemon_type = "Psychic"
            elif pokemon_name == "cubone":
                pokemon_type = "Ground"
            elif pokemon_name == "eevee":
                pokemon_type = "Normal"
            elif pokemon_name == "snorlax":
                pokemon_type = "Normal"
            elif pokemon_name == "dragonite":
                pokemon_type = "Dragon / Flying"
            elif pokemon_name == "mewtwo":
                pokemon_type = "Psychic"
            else:
                pokemon_type = "Unknown"
            if pokemon_type not in pokemon_data_by_type:
                pokemon_data_by_type[pokemon_type] = {}
            pokemon_data_by_type[pokemon_type][pokemon_name] = {
                "abilities": abilities,
                "weight": weight
            }
            
    for pokemon_type, pokemon_info in pokemon_data_by_type.items():
        print(f"{pokemon_type}:")
        for name, info in pokemon_info.items():
            print(f"  {name}: Abilities: {info['abilities']}, Weight: {info['weight']}")

if __name__ == "__main__":
    main()
```
